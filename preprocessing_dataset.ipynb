```python
# importing libraries

import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Count the jobs for each company and display that in order 
demanding_companies = pd.DataFrame(df.loc[:,"Company"].value_counts().sort_values(ascending=False))
demanding_companies
```

```python
# Print skills one by one and how many each repeated and order the output to find out the most important skills required
# we can solve this problem using two ways:
# 1st way:

skills=[]
for skill in df.loc[:,'Skills'].str.split(','):
    skills+=[skill1.strip() for skill1 in skill]
```

```python
# print(skills)
```

```python
skills_dic={}
for skill in skills:
    skills_dic[skill]=skills_dic.get(skill,0)+1
    
```

```python
# print(skills_dic)
```

```python
sorted_skills= dict(sorted(skills_dic.items(), key= lambda item: item[1], reverse=True))
```
